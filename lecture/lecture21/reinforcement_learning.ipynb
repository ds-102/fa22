```python
def fib(n):
    print(f"Computing fib({n})")
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
```

Memoization

```python
fib_numbers = {0: 0, 1: 1}
def fib_memo(n):
    if n in fib_numbers:
        print(f"Loading saved fib_memo({n})")
        return fib_numbers[n]
    else:
        print(f"Computing fib_memo({n})")
        answer = fib_memo(n-1) + fib_memo(n-2)
        fib_numbers[n] = answer
        return answer
```

## Road trip



* We have locations $0$-$n$ along our drive (we start at 0) and want to reach n
* Our gas tank holds $k$ units of gas
* Each location $i$ from 0 to $n-1$ sells gas in 1-unit increments for  $\$c_i$ per unit (known)
* Moving forward uses up 1 unit of gas
* Goal: get to location $n$ as cheaply as possible

```python
gas_prices = np.array([10, 8, 7, 2, 3, 5])
n = len(gas_prices)  # index of last location
gas_capacity = 3

def compute_trip_cost(current_gas, location):
    if current_gas < 0 or current_gas > gas_capacity:
        return np.inf
    if location == n:
        return 0
    print(f"Computing cost with {current_gas} gallons at location {location}")
    cost_to_fill = compute_trip_cost(current_gas + 1, location) + gas_prices[location]
    cost_to_drive = compute_trip_cost(current_gas - 1, location + 1)
    return min(cost_to_fill, cost_to_drive)
```
